```python
import wikipedia as wp
```

```python
# Let's search the first result of an "EPFL" query

an = wp.search("EPFL")[0]
pg = wp.page(an)
pg.title
```

```python
# Its URL

pg.url
```

```python
# Its 100 first words

" ".join(pg.content.split(" ")[:100])
```

```python
# The links found in the page

pg.links
```

```python
# How about the 10 furst sentences in the French article in French?

wp.set_lang("fr")
pg = wp.search(an)
wp.summary(pg, sentences=6)
```

```python
# Now let's see how many articles we can get for a query

occurences = wp.search("EPFL", results=10000)
print("wikipedia.search only returns a maximum of", len(occurences), "occurences even when asked for 10'000.")
```

Python's `wikipedia` module is not intended for big data uses as seen above. The `wikipedia.search` function will only return the first 500 occurences of a word, which is not appropriate. We will thus use Pywikibot from https://pywikibot.toolforge.org/
